```python
#Converting or encoding the Time_of_Day and Location column using Label encoder (ordinal_data) and one_hot_encoder(nominal data)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Initialize the LabelEncoder
label_encoder = LabelEncoder()
```

```python
Time_of_Day_encoded = label_encoder.fit_transform(bank_data['Time_of_Day'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder =  OneHotEncoder(sparse=False)
```

```python
Location_encoded = one_hot_encoder.fit_transform(bank_data[['Location']])
```

```python
bank_data['Time_of_Day_encoded'] = Time_of_Day_encoded
```

```python
Location_encoded_df = pd.DataFrame(Location_encoded, columns=one_hot_encoder.get_feature_names_out(['Location']))
bank_data = pd.concat([bank_data, Location_encoded_df], axis=1)
```

```python
bank_data
```

```python
#dropping the original columns and last NAN row
bank_data_cleaned = bank_data.drop(columns=['Location','Time_of_Day'], axis=1)
```

```python
bank_data_cleaned = bank_data_cleaned.dropna()
```

```python
bank_data_cleaned
```

```python
bank_data_cleaned['Is_Fraudulent'].value_counts()
```

```python
#separating datasets
legit = bank_data_cleaned[bank_data_cleaned['Is_Fraudulent']==0]
fraud = bank_data_cleaned[bank_data_cleaned['Is_Fraudulent']==1]
```

```python
fraud.shape
```

```python
legit.shape
```

```python
#Undersampling
legit_sample = legit.sample(n=8)
```

```python
new_data = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_data
```

```python
new_data['Is_Fraudulent'].value_counts()
```

```python
#Spliting the dataset and feature
X= new_data.drop(columns='Is_Fraudulent', axis=1)
y= new_data['Is_Fraudulent']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#training the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=102)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
#Model Evaluation

#Accuracy Testing

X_train_prediction = lr.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = lr.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
